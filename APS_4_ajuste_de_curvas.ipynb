### Utilizando uma regress√£o linear,

```python
#pontos dados
x = [0,0.125,0.25,0.375,0.5]
y = [0,6.24,7.75,4.85,0]



#c√°lculo da regress√£o
xy = 0
x2 = 0
sx = 0
sy = 0
for i in range(5):
    xy += x[i]*y[i]
    sx += x[i]
    sy += y[i]
    x2 += x[i]**2
    
a1 = (5*xy - (sx*sy))/((5*x2) - (sx**2))
a0 = (sum(y)/5) - a1*(sum(x)/5)



#gr√°fico 
plt.figure(figsize=(8,8))
ajuste = [a0 + a1*i for i in x]
plt.plot(x, ajuste, color="black")
plt.title("i (A) vs. t (s)")
plt.grid()
```

```python
print(a0+a1*0.23)
```

## Em 0,23s, a corrente √© de 3.79024 A.
