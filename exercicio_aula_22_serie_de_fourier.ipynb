## Gráfico geral - pontos experimentais da tabela e graus 2, 3, 4, 5 e 6

```python
# A área será computada usando a regra de Simpson. dx indica o espaçamento entre os dados no eixo x (tempos)

coeffs = []
areas = []
r_s = [] # r quadrado

plt.figure(figsize=(20,20))
plt.scatter(t,c, color='blue',linewidth=4.0)


# GRAU 2
deg = np.polyfit(t,c,2)
f = np.linspace(0,24,100).tolist()
re = [deg[0]*(i**2) + deg[1]*i + deg[2] for i in f]
plt.plot(f,re, color='purple',linewidth=1.5,label='grau 2, r^2 = 0.8819')
coeffs.append(deg)
area = simps(re, dx=0.24)
areas.append(area)
#cálculo de r ao quadrado
curve = [deg[0]*(i**2) + deg[1]*i + deg[2] for i in t]
r_squared = (np.corrcoef(c, curve)[0,1])**2
r_s.append(r_squared)


# GRAU 3
deg = np.polyfit(t,c,3)
f = np.linspace(0,24,100).tolist()
re = [deg[0]*(i**3) + deg[1]*(i**2) + deg[2]*i + deg[3] for i in f]
plt.plot(f,re, color='red',linewidth=1.5,label='grau 3, r^2 = 0.9037')
coeffs.append(deg)
area = simps(re, dx=0.24)
areas.append(area)
#cálculo de r ao quadrado
curve = [deg[0]*(i**3) + deg[1]*(i**2) + deg[2]*i + deg[3] for i in t]
r_squared = (np.corrcoef(c, curve)[0,1])**2
r_s.append(r_squared)


# GRAU 4
deg = np.polyfit(t,c,4)
f = np.linspace(0,24,100).tolist()
re = [deg[0]*(i**4) + deg[1]*(i**3) + deg[2]*(i**2) + deg[3]*i + deg[4] for i in f]
plt.plot(f,re, color='green',linewidth=1.5,label='grau 4, r^2 = 0.9970')
coeffs.append(deg)
area = simps(re, dx=0.24)
areas.append(area)
#cálculo de r ao quadrado
curve = [deg[0]*(i**4) + deg[1]*(i**3) + deg[2]*(i**2) + deg[3]*i + deg[4] for i in t]
r_squared = (np.corrcoef(c, curve)[0,1])**2
r_s.append(r_squared)


# GRAU 5
deg = np.polyfit(t,c,5)
f = np.linspace(0,24,100).tolist()
re = [deg[0]*(i**5) + deg[1]*(i**4) + deg[2]*(i**3) + deg[3]*(i**2) + deg[4]*i + deg[5] for i in f]
plt.plot(f,re, color='magenta',linewidth=2.0 ,label='grau 5, r^2 = 0.9991')
coeffs.append(deg)
area = simps(re, dx=0.24)
areas.append(area)
#cálculo de r ao quadrado
curve = [deg[0]*(i**5) + deg[1]*(i**4) + deg[2]*(i**3) + deg[3]*(i**2) + deg[4]*i + deg[5] for i in t]
r_squared = (np.corrcoef(c, curve)[0,1])**2
r_s.append(r_squared)


# GRAU 6
deg = np.polyfit(t,c,6)
f = np.linspace(0,24,100).tolist()
re = [deg[0]*(i**6) + deg[1]*(i**5) + deg[2]*(i**4) + deg[3]*(i**3) + deg[4]*(i**2) + deg[5]*i + deg[6] for i in f]
plt.plot(f,re, color='black',linewidth=1.5 ,label='grau 6, r^2 = 0.9995')
coeffs.append(deg)
area = simps(re, dx=0.24)
areas.append(area)
#cálculo de r ao quadrado
curve = [deg[0]*(i**6) + deg[1]*(i**5) + deg[2]*(i**4) + deg[3]*(i**3) + deg[4]*(i**2) + deg[5]*i + deg[6] for i in t]
r_squared = (np.corrcoef(c, curve)[0,1])**2
r_s.append(r_squared)


#linha do zero
a = np.linspace(0,24).tolist()
b = [0 for i in a]
plt.plot(a,b, color='pink',linewidth=1.5, label = 'zero')


plt.title('Gráficos - Concentração de Corante vs. tempo')
plt.ylabel('c (mg/L)')
plt.xlabel('t (s)')
plt.legend()
plt.grid()
plt.show()
```
