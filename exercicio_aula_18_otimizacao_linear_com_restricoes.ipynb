## No gráfico abaixo, as linhas pontilhadas são restrições e a linha cheia é a função a ser maximizada (neste caso, o lucro).
## A linha pontilhada laranja é a restrição do estoque de matéria-prima.

## Para iniciar o problema, será assumido um lucro de 10k reais para analisar o comportamento. Assim, tem-se o seguinte gráfico:

```python
def f1(x):
    return 550 - x
def f2(x):
    return (40 - 0.04*x)/0.12
def f3(x):
    return (9500 - 20*x)/5
def f4(x):
    return 0
def lucro(x):
    return (10000-45*x)/20

x = Symbol('x')
x1, =  solve(f1(x)-f2(x))
x2, =  solve(f1(x)-f3(x))
x3, =  solve(f2(x)-f3(x))
x4, =  solve(f3(x)-f4(x))
x5 = 0
x6 = 0

y1 = f1(x1)
y2 = f1(x2)
y3 = f2(x3)
y4 = f3(x4)
y5 = 0
y6 = 333.333

plt.figure(figsize=(10,10))

plt.plot(x1,f1(x1),'go',markersize=5)
plt.plot(x2,f1(x2),'go',markersize=5)
#plt.plot(x3,f2(x3),'go',markersize=5)
plt.plot(x4,f4(x4),'go',markersize=5)
plt.plot(x5,y5,'go',markersize=5)
plt.plot(x6,y6,'go',markersize=5)

plt.fill([x1,x2,x3,x4,x5,x6,x1],[y1,y2,y3,y4,y5,y6,y1],'pink',alpha=0.5)

xr = np.linspace(0.5,550,100)
y1r = f1(xr)
y2r = f2(xr)
y3r = f3(xr)
ylucror = lucro(xr)

plt.plot(xr,y1r,'k--', color='yellowgreen', label='tempo útil para produção')
plt.plot(xr,y2r,'k--', color='mediumslateblue', label='armazenamento')
plt.plot(xr,y3r,'k--', color='darkorange', label='matéria-prima')
plt.plot(xr,ylucror, color='k', label='lucro')

plt.xlim(0,500)
plt.ylim(0,550)

plt.legend()
plt.show()
```

# O lucro é, neste gráfico, representado pela linha cheia. Desta forma, variando-se o valor do COEFICIENTE LINEAR DA RETA PRETA CHEIA (lucro) até encontrar a intersecção representada no gráfico com o ponto ótimo (que é aquele ponto ali mais à direita, onde duas retas pontilhadas se intersectam). Assim, pelo método gráfico, é possível concluir que o lucro máximo é de 22250 conforme o gráfico ABAIXO.

```python
def f1(x):
    return 550 - x
def f2(x):
    return (40 - 0.04*x)/0.12
def f3(x):
    return (9500 - 20*x)/5
def f4(x):
    return 0
def lucro(x):
    return (22250-45*x)/20

x = Symbol('x')
x1, =  solve(f1(x)-f2(x))
x2, =  solve(f1(x)-f3(x))
x3, =  solve(f2(x)-f3(x))
x4, =  solve(f3(x)-f4(x))
x5 = 0
x6 = 0

y1 = f1(x1)
y2 = f1(x2)
y3 = f2(x3)
y4 = f3(x4)
y5 = 0
y6 = 333.333

plt.figure(figsize=(10,10))

plt.plot(x1,f1(x1),'go',markersize=5)
plt.plot(x2,f1(x2),'go',markersize=5)
#plt.plot(x3,f2(x3),'go',markersize=5)
plt.plot(x4,f4(x4),'go',markersize=5)
plt.plot(x5,y5,'go',markersize=5)
plt.plot(x6,y6,'go',markersize=5)

plt.fill([x1,x2,x3,x4,x5,x6,x1],[y1,y2,y3,y4,y5,y6,y1],'pink',alpha=0.5)

xr = np.linspace(0.5,550,100)
y1r = f1(xr)
y2r = f2(xr)
y3r = f3(xr)
ylucror = lucro(xr)

plt.plot(xr,y1r,'k--', color='yellowgreen', label='tempo útil para produção')
plt.plot(xr,y2r,'k--', color='mediumslateblue', label='armazenamento')
plt.plot(xr,y3r,'k--', color='darkorange', label='matéria-prima')
plt.plot(xr,ylucror, color='k', label='lucro')

plt.xlim(0,500)
plt.ylim(0,550)

plt.legend()
plt.show()
```

# TRANSFORMANDO ESSA MATRIZ EM MATRIZ IDENTIDADE POR GAUSS JORDAN, É POSSIVEL DETERMINAR S1, S2 e S3 e depois, por fim, A e B.

```python
#código para pivotar a matriz inteira
for i in range(1,4):
    for j in range(3):
        if i > j:
            if matrix[i][j] != 0:
                temp = matrix[i][j]/matrix[j][j]
                for k in range(4):
                    matrix[i][k] -= temp*matrix[j][k]
```

```python
matrix = np.transpose(matrix)
```
