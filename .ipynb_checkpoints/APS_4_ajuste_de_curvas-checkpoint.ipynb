## (a)

```python
#pontos dados
x = [2,3,4,5,7,10]
y = [5.2,7.8,10.7,13,19.3,27.5]



#cálculo da regressão
xy = 0
x2 = 0
sx = 0
sy = 0
for i in range(6):
    xy += x[i]*y[i]
    sx += x[i]
    sy += y[i]
    x2 += x[i]**2
    
a1 = (6*xy - (sx*sy))/((6*x2) - (sx**2))
a0 = (sum(y)/6) - a1*(sum(x)/6)



#gráfico 
plt.figure(figsize=(8,8))
ajuste = [a0 + a1*i for i in x]
plt.plot(x, ajuste, color="black")
plt.title("i (A) vs. V (V)")
plt.grid()
```

```python
print(a0 + a1*3.5)
```

### Para 3,5V, a corrente é de 9.2363 A.

```python
x = [200,250,300,375,425,475,600]
y = [7.5,8.6,8.7,10,11.3,12.7,15.3]
```

```python
plt.figure(figsize=(6,6))
plt.grid()
plt.plot(x,y,'o', color='black')
plt.plot(x,y, color='black')
```

```python
xy = 0
x2 = 0
sx = 0
sy = 0
for i in range(7):
    xy += x[i]*y[i]
    sx += x[i]
    sy += y[i]
    x2 += x[i]**2
    
a1 = (7*xy - (sx*sy))/((7*x2) - (sx**2))

a0 = (sum(y)/7) - a1*(sum(x)/7)
```

```python
plt.figure(figsize=(8,8))
ajuste = [a0 + a1*i for i in x]
plt.plot(x, ajuste, color="black")
plt.title("deformação % vs T(°C)")
plt.grid()
```

```python
print(a0 + a1*400)
```

## Para uma temperatura de 400, a deformação é de 11.07212%.
