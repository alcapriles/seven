```python
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

# Primeiro caso: passo de h = 0.25.

```python
# número de termos: 
h = 0.25
print(5/h)
```

```python
# colocando dois termos a mais no início e no fim para o cálculo da derivada segunda:

intervalo = [i*h - (2 + 2*h)  for i in range(21)]
print(intervalo)
```

```python
#valores reais para a função:
f = [x**3 -(2*x) + 4 for x in intervalo]
#valores reais para a primeira derivada da função:
print(f)
f1 = [3*(x**2) - 2 for x in intervalo]
print(f1)
#valores reais para a segunda derivada da função:
f2 = [6*x for x in intervalo]
print(f2)
```

## Primeira Derivada

```python
f1_forward = []
y = 0
for i in range(2,19):
    y = (f[i+1] - f[i])/h
    f1_forward.append(y)
    
print(f1_forward) 
print(len(f1_forward))
```

```python
f1_backwards = []
y = 0
for i in range(2,19):
    y = (f[i] - f[i-1])/h
    f1_backwards.append(y)
print(f1_backwards) 
print(len(f1_backwards))
```

```python
y = 0
f1_centered = []
for i in range(2,19):
    y = (f[i+1] - f[i-1])/(2*h)
    f1_centered.append(y)
print(f1_centered) 
print(len(f1_centered))
```

## Segunda derivada:

```python
y = 0
f2_forward = []
for i in range(2,19):
    y = (f[i+2] - 2*f[i+1] +f[i])/(h**2)
    f2_forward.append(y)
    
print(f2_forward) 
print(len(f2_forward))
```

```python
y = 0
f2_backwards = []
for i in range(2,19):
    y = (f[i] - 2*f[i-1] +f[i-2])/(h**2)
    f2_backwards.append(y)
print(f2_backwards) 
print(len(f2_backwards))
```

```python
y = 0
f2_centered = []
for i in range(2,19):
    y = (f[i+1] - 2*f[i] +f[i-1])/(h**2)
    f2_centered.append(y)
print(f2_centered) 
print(len(f2_centered))
```

## Gráficos:

```python
intervalo = [-2.0, -1.75, -1.5, -1.25, -1.0, -0.75, -0.5, -0.25, 0.0, 0.25, 0.5, 0.75, 1.0, 1.25, 1.5, 1.75, 2.0]
f1 = [3*(x**2) - 2 for x in intervalo]
f2 = [6*x for x in intervalo]

#primeira derivada:
plt.figure(figsize=(13,10))
plt.plot(intervalo, f1, color="black")

plt.plot(intervalo, f1_forward, color="blue")
plt.plot(intervalo, f1_backwards, color="red")
plt.plot(intervalo, f1_centered, color="magenta")

plt.show()
```

```python
#segunda derivada:
plt.figure(figsize=(13,10))
plt.plot(intervalo, f2, color="black")

plt.plot(intervalo, f2_forward, color="pink")
plt.plot(intervalo, f2_backwards, color="yellow")
plt.plot(intervalo, f2_centered, color="brown")

plt.show()
```

## Comentário


É possível obervar que, ao diminuir-se o passo, diminui-se o erro de truncamento, ou seja, as linhas coloridas ficam mais próximas no gráfico da linha
preta, a qual representa os valores reais da função.
Em outras palavras, no caso em que o passo é de 0.1, os valores estimados estão mais próximos dos valores reais, ou seja, o erro é menor em relação ao caso em que o passo é de 0.25.
Também se nota que o cálculo por meio da diferença dividida centrada é o que possui o menor erro, pelo fato de a curva (magenta) estar quase sobreposta à curva real, representada nos gráficos pela linha preta.
No gráfico da derivada segunda, estas duas linhas se encontram sobrepostas.
