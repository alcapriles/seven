```python
import math
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
#parâmetros do problema:
a = 1250 #área (metros quadrados)
q = 450 #vazão (metros cúbicos por dia)
alpha = 150
```

## Primeiro caso (step = 0.5)

```python
t = np.arange(0,10.5,0.5).tolist()
print(len(t))
print(t)
```

```python
y_values = [0]
y= 0
for i in range(len(t)-1):
    y += 0.5*(   3*(q/a)*(math.sin(t[i])**2) - ( alpha*( (1+y)**1.5) )/a    )
    
    y_values.append(y)
    
print(y_values)
```

```python
print(len(y_values))
```

## Segundo caso (step=0.1)

```python
tt = np.arange(0,10.1,0.1).tolist()
```

```python
y_values_caso_dois = [0]
y = 0
for i in range(len(tt)-1):
    y += 0.1*(   3*(q/a)*(math.sin(tt[i])**2) - ( alpha*( (1+y)**1.5) )/a    )
    y_values_caso_dois.append(y)
```

## Terceiro caso (step=0.01)

```python
ttt = np.arange(0,10.01,0.01).tolist()
```

```python
y_values_caso_tres = [0]
y = 0
for i in range(len(ttt)-1):
    y += 0.01*(   3*(q/a)*(math.sin(ttt[i])**2) - ( alpha*( (1+y)**1.5) )/a    )
    y_values_caso_tres.append(y)
```

## TODOS OS GRÁFICOS

```python
plt.figure(figsize=(10,10))
plt.plot(t, y_values, color='magenta')
plt.plot(tt,y_values_caso_dois, color='orange')
plt.plot(ttt,y_values_caso_tres, color='black')
plt.title('Profundidade x Tempo')
plt.xlabel("tempo (d)")
plt.ylabel("y (m)")
plt.grid()
plt.show()
```

## Comentário
O gráfico do caso 2 está mais próximo da idealidade do modelo físico do que o gráfico do caso 1. Isto se dá porque, no caso 2, 
a distância entre cada passo da iteração é 5 vezes menor que no caso 1, o que faz com que os cálculos demorem 5 vezes mais para
serem processados, porém torna o resultado mais preciso.

Já entre os casos com passos de 0.001 e 0.5, a diferença é muito pequena, levantando a questão de se vale realmente a pena ter estes passos a mais com mais esforço computacional.
