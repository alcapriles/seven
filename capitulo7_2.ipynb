```python
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
#the function must necessarily take only one argument
def evaluate(range_for_evaluation, function):
    ordinate = []
    for i in range_for_evaluation:
        ordinate.append(function(i))
    return ordinate
```

```python
# FOR ZOOMING IN in the graph: plt.axis([xmin,xmax,ymin,ymax]) ----------- (zoom range)
def plot(x,y, title, xlabel, ylabel, zoom, zoom_range):
    a = 0
    plt.figure(figsize=(10,10)) #(I need to find a nice size)
    plt.plot(x,y, color='magenta')
    if zoom == True:
        plt.axis(zoom_range)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.axhline(y=0, color='k') #this makes the x axis appear
    plt.grid()
    plt.show()
    return a
```

```python
#function for a table:
def table(index, table_values, title):
    table = pd.DataFrame(table_values)
    table.index = [i for i in index]
    table.columns = [title]
    return table
```

## Método de Muller:

Os inputs desse programa serão x0, x1 e x2, que serão os "chutes" iniciais para as raízes.

## Maior raiz:

```python
#inputs:
x0 = 2.5
x1 = 3
x2 = 2.1

error = 0.5
#contador de iterações:
i = 0
desired_error = 4*(10**-4)

while error > desired_error:
#em seguida, os parâmetros h e delta do método são determinados:
    h0 = x1 - x0
    h1 = x2 - x1
    
    fx1 = x1**3 + x1**2 - 4*x1 - 4
    fx0 = x0**3 + x0**2 - 4*x0 - 4
    fx2 = x2**3 + x2**2 - 4*x2 - 4

    delta0 = (fx1-fx0)/(x1-x0)
    delta1 = (fx2-fx1)/(x2-x1)
    
    #em seguida, os parâmetros a, b e c do método são determinados:
    a = (delta1 - delta0)/(h1 + h0)
    b = a*h1 + delta1
    c = fx2
    
    #x3 é determinado:
    
    
    v11 = b+(math.sqrt(-4*a*c + b**2))
    v22 = b-(math.sqrt(-4*a*c + b**2))
    
    v1 = math.fabs(v11)
    v2 = math.fabs(v22)
    
    if v1 > v2:
        x3 = x2 + (-2*c)/(b+(math.sqrt(-4*a*c + b**2)))
    if v1 < v2:
        x3 = x2 + (-2*c)/(b-(math.sqrt(-4*a*c + b**2)))
    if v1 == v2:
        x3 =  x2 + (-2*c)/b

    #o erro é computado para checar se haverá mais alguma iteração:
    error = math.fabs(100*(x3 - x2)/x3)
    
    x0 = x1
    x1 = x2
    x2 = x3
    
    i+= 1

print(x3)
print(error)  
print(i)
```

## Assim, para a maior raiz e para os chutes iniciais de x0 = 2.5, x1 = 3, e x2 = 2.1, foram feitas 4 iterações, o erro foi de 6.837537203097099e-08
## e o valor final encontrado para a raiz foi igual a 2.0.
